```python
import pandas as pd

df = pd.read_csv("C:/Users/moura/PAD_JOB/TextBlob/dataset-Traduzido.csv")


display(df)
```

```python
#df = df.drop('TextBlob', axis=1)
df = df.drop('Unnamed: 6', axis=1)
# Substituir todos os NaN por 0
df['TextBlob'] = df['TextBlob'].fillna(0)
display(df)
```

```python
from textblob import TextBlob
cont = 0
polaridade_lista = []



for index, row in df.iterrows():
  #Peque a coluna que tem o texto traduzido e coloque em row["nome da coluna"]
    frase1 = row['texto_traduzido']  
    cont += 1
    
    tb = TextBlob(frase1)
    #tb_en = tb.translate(from_lang='pt_br', to='en')
    #polarity = tb_en.sentiment.polarity
    
    polarity = tb.sentiment.polarity
    polaridade_lista.append(polarity)
    
df['TextBlob'] = polaridade_lista
    
#print(f"Frase {cont}: {frase1}")
#print(f"Polaridade: {polarity}")
#print("---")
    
```
