```python
import requests
import json

# Defina a chave da API
api_key = "AIzaSyB93P1qS8i3S9EO66u9gs8cMCI2ivE6wsM"

# Defina a url base
url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json?"

# Defina a consulta de pesquisa
query = "bars in Goinia"

# Faa a solicitao
response = requests.get(url + 'input=' + query +
                        '&inputtype=textquery&fields=photos,formatted_address,name,rating,opening_hours,geometry,place_id' +
                        '&key=' + api_key)

# Formate a resposta como JSON
x = response.json()

# Se a pesquisa tiver resultados
if x['status'] == 'OK':
    # Para cada resultado
    for i in range(len(x['candidates'])):
        # Obtenha os detalhes do local
        place_id = x['candidates'][i]['place_id']
        fields = 'review'
        details_url = f"https://maps.googleapis.com/maps/api/place/details/json?place_id={place_id}&fields={fields}&key={api_key}"
        details_response = requests.get(details_url)
        details = details_response.json()

        print("Nome: ", x['candidates'][i]['name'])
        print("Endereo: ", x['candidates'][i]['formatted_address'])
        print("Avaliao: ", x['candidates'][i]['rating'])
        print("Comentrios: ")

        if details['status'] == 'OK':
            for review in details['result']['reviews']:
                print("Autor: ", review['author_name'])
                print("Avaliao: ", review['rating'])
                print("Comentrio: ", review['text'])
                print("Data: ", review['relative_time_description'])
                print("\n")
else:
    print("Sem resultados")
```

```python
import requests
from bs4 import BeautifulSoup

# Get the Google Maps search results page for "restaurantes em So Paulo"
response = requests.get("https://www.google.com/maps/search/restaurantes+em+So+Paulo")


# Parse the HTML response with BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
print(soup)

# Find all the restaurant results
restaurants = soup.find_all("div", class_="from_the_business")
print(restaurants)

# Print the name, address, and rating of each restaurant
for restaurant in restaurants:
    name = restaurant.find("div", class_="business-name").text
    address = restaurant.find("div", class_="business-address").text
    rating = restaurant.find("div", class_="rating").text
    print(f"Name: {name}\nAddress: {address}\nRating: {rating}")
```
