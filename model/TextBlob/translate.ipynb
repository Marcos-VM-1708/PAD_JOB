# Polaridade de uma frase


```python
import pandas as pd

frase = pd.read_csv("C:/Users/moura/PAD_JOB/TextBlob/BDTeste-2.csv")


display(frase)
```

```python
tb = frase.loc[4, 'COMENTÁRIO']
print(tb)
```

```python
from textblob import TextBlob

tb = TextBlob(tb)
tb_en = tb.translate(from_lang='pt_br', to='en')
print(tb_en)
```

```python
polarity = tb_en.sentiment.polarity
print(f"Polaridade: {polarity}")
```

# Tradução dos comentários

```python
import pandas as pd

df = pd.read_csv("C:/Users/moura/PAD_JOB/TextBlob/BDTeste-2.csv")


display(frase)
```

```python
import pandas as pd
from textblob import TextBlob

# Suponha que este é o seu DataFrame original
#data = {'texto': ['Olá mundo', 'Como você está?', 'Estou bem']}
#df = pd.DataFrame(data)

# Função para traduzir texto
def traduzir(texto):
    try:
        return str(TextBlob(texto).translate(from_lang='pt_br', to='en'))
    except:
        return texto


# Aplicar a função de tradução para a coluna 'texto'
df['texto_traduzido'] = df['COMENTÁRIO'].apply(traduzir)

# Agora 'df' tem uma nova coluna 'texto_traduzido' com o texto traduzido
print(df)

# Para armazenar a coluna traduzida em outra tabela (novo DataFrame)
novo_df = df['texto_traduzido'].to_frame()
print(novo_df)
```
